```python
image = np.load('../examples/CuGB.npy')
```

```python
from sklearn.decomposition import PCA
```

```python
patch_x_all = [2,8,16,32]
pca = PCA(10)
evr_all = []
for patch_x in patch_x_all:
    step = 5
    descriptor_name = 'local_correlation_map'
    seg = segmentationSTEM( n_patterns=3,
                           window_x=20,window_y=20,
                           patch_x=patch_x,patch_y=patch_x,
                           step=step,
                           descriptor_name=descriptor_name,
                           method='direct',
                           upsampling=True)
    descriptors = seg.get_descriptors(image)
    shape = descriptors.shape
    pca.fit(np.reshape(descriptors, (-1, shape[-1])))
    evr_all.append(pca.explained_variance_ratio_)
```

```python
plot_is = [2,]

for i,evr in enumerate(evr_all):
    plt.plot( [0]+list(np.cumsum(evr)),'-o',label=str(patch_x_all[i]*2))
plt.plot([0,10],[0.9,0.9],'--',color='black')
plt.legend(loc='center right')
plt.xlabel('# of PCA features')
plt.ylabel('Percentage of explained variance')

plt.xlim(0,10)
plt.ylim(0,1)
```
